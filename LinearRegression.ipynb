```python
#importing modules 
import random 
import matplotlib.pyplot as plt
import math
```

```python
# inputs required 
initial = int(input())
m = int(input())
c = int(input())
n = int(input())
alpha = int(input())
beta = float(input())
gamma = float(input())
```

```python
# error funtion 
def error(alpha, initial, n):
  """
  Returns: It returns list of error 
  
  Parameter alpha: no to increase error 
  Precondition : It must be integer 

  parameter initial: It is the initial value for forloop
  Precondition : it must be integer

  parameter n: no. of observations 
  precondition : it must be integer
  """
  epsilon = []
  for i in range(initial,n+1):
    ep = (random.random() - 0.5) * alpha
    epsilon.append(ep)
  return epsilon
epsilon = error(alpha, initial, n)
```

```python
#x list 
x_list = [i for i in range(1,n+1)]
```

```python
  
y_list = []
for i in range(n):
  y = beta * math.exp(gamma * x_list[i] + epsilon[i])
  y_list.append(y)
```

```python
z_list = []
for i in range(n):
  log_value = math.log(y_list[i])
  z_list.append(log_value)
```

```python
# def minimising(x_list,z_)
slope = 0
x_z = 0
x_avg = 0
z_avg = 0
x_sq_bar = 0
x_bar_sq = 0

for i in range(n):
  x_z += (x_list[i] * z_list[i])
  x_avg += x_list[i]
  z_avg +=z_list[i]
  x_sq_bar +=(x_list[i])**2
  x_bar_sq +=(x_list[i])
slope = ((x_z/n) - (x_avg/n) * (z_avg/n)) / ((x_sq_bar/n) - ((x_bar_sq/n)**2))

constant = (z_avg/n) - (slope * (x_avg/n))
bita = math.exp(constant)
gmma = slope
```

```python
y_new_list = []
for i in range(n):
  y_i = bita * math.exp(gmma*x_list[i])
  y_new_list.append(y_i)
```

```python
# Plotting the first scatter plot
plt.plot(x_list, y_list, label='Linear Regression 1')

# Plotting the second scatter plot
plt.plot(x_list, y_new_list, label='Linear Regression 2')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

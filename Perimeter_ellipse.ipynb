```python
from scipy.integrate import quad
import math
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
i = 2.0
list_a =[]
list_b =[]
while i <= 4:
  list_a.append(i)
  i += 0.1
print(len(list_a))
j = 1.0

while j <= 2:
    list_b.append(2)
    j += 0.1
list_b_original = []

for i in range(2):
  list_b_original+=list_b[:]
print(len(list_b_original))

result = []
result_two = []
result_three = []
result_four = []
result_five = []
for i in range(len(list_a)):
    def f(t):
        return math.sqrt(list_a[i]**2*math.sin(t)**2 + list_b_original[i]**2*math.cos(t)**2)
    x,y= quad(f,0, 2*math.pi)
    result.append(x)
    result_two.append(math.pi*(3*(list_a[i] + list_b_original[i]) - math.sqrt((3*list_a[i] + list_b_original[i])*(list_a[i]+ 3*list_b_original[i]))))
    result_three.append(math.pi*(((53*list_a[i])/3)+((717*list_b_original[i])/35) - math.sqrt(269*(list_a[i]**2) + 667 * list_a[i]*list_b_original[i]+ 371 * (list_b_original[i])**2)))
    result_four.append(math.pi*((6*list_a[i])/5 + (3*list_b_original[i])/4))
    result_five.append(2*math.pi*(math.sqrt((list_a[i]**2 + list_b_original[i]**2)/2)))


```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(list_a, list_b_original, result, 'red')
# Add labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Plot')
# ShowÂ plot
plt.show()
```

```python
# Create figure and axes
fig, ax = plt.subplots()
# Plot data
ax.plot(list_a, result, label='result')
ax.plot(list_a, result_two, label='result_two')
ax.plot(list_a, result_three, label='result_three')
ax.plot(list_a, result_four, label='result_four')
ax.plot(list_a, result_five, label='result_five')

# Add legend and labels
ax.legend()
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Multiple Line Plot')

# Show plot
plt.show()
```

```python
p_result_two = []
p_result_three = []
p_result_four = []
p_result_five = []
for i in range(len(list_a)):
  p_result_two.append(abs(result[i]-result_two[i])/result[i] * 100)
  p_result_three.append(abs(result[i]-result_three[i])/result[i] * 100)
  p_result_four.append(abs(result[i]-result_four[i])/result[i] * 100)
  p_result_five.append(abs(result[i]-result_five[i])/result[i] * 100)
  
```

```python
fig, ax = plt.subplots()

# Plot data
ax.plot(list_a, p_result_two, label='percent_result_two')
ax.plot(list_a, p_result_three, label='percent_result_three')
ax.plot(list_a, p_result_four, label='percent_result_four')
ax.plot(list_a, p_result_five, label='percent_result_five')

# Add legend and labels
ax.legend()
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Multiple Line Plot')

# Show plot
plt.show()
```

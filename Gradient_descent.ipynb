```python
#gradient descent for calculating minima f(x) = sin(x)
ita = 0.01
xn_initial = 1.0 #start value 

variable = True
while variable:

  if gradient_func(xn_initial) > 0:
    xn_initial -= 0.1 #if the slope positive we need to move left
    xn_two = xn_initial - ita * (gradient_func(xn_initial))
    if gradient_func(xn_two) < 0:
      print(xn_initial/np.pi) #minima
      variable = False 
  elif  gradient_func(xn_initial) < 0:
    xn_initial += 0.1 #when slope negative need to move right
    xn_two = xn_initial - ita * (gradient_func(xn_initial))
    if gradient_func(xn_two) > 0:
      print(xn_initial/np.pi) #minima
      variable = False  

##gradient descent for calculating maxima f(x) = sin(x)
ita = 0.01
xn_initial = 1.0 #start value 
variable = True

while variable:
  if gradient_func(xn_initial) > 0:
    xn_initial += 0.1 #if the slope positive we need to move right
    xn_two = xn_initial + ita * (gradient_func(xn_initial))

    if gradient_func(xn_two) < 0:
      print(xn_initial/np.pi) #maxima
      variable = False 

  elif  gradient_func(xn_initial) < 0:
    xn_initial -= 0.1 #when slope negative need to move left
    xn_two = xn_initial + ita * (gradient_func(xn_initial))

    if gradient_func(xn_two) > 0:
      print(xn_initial/np.pi) #minima
      variable = False  
```
